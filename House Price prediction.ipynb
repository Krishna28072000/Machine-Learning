```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("E:\\Machine Learning\\data.csv")
```

```python
df.head()
```

```python
house_size = df['sqft_living'].values
house_price = df['price'].values
```

```python
plt.scatter(house_size, house_price, marker = 'o', color = 'blue')
plt.xlabel('House Size (sqft)')
plt.ylabel('House Price ($)')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(house_size, house_price, test_size = 0.2, random_state = 42)
x_train = X_train.reshape(-1, 1)
x_test = X_test.reshape(-1,1)

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
predictions = model.predict(x_test)

plt.scatter(x_test, y_test, marker = 'o', color = 'red', label = 'Actual Prices')
plt.plot(x_test, predictions, color = 'blue', linewidth = 2, label = 'Predicted Prices')
plt.title('House Price Prediction with Linear Regression')
plt.xlabel('House Price (sq.ft)')
plt.ylabel('House Price($)')
plt.legend()
plt.show()
```
