**4. Boxplot**

* A box plot in terms of categorical columns visualizes the distribution of a continuous variable across different categories. It shows the median, quartiles, and potential outliers for each category. Box plots are useful for comparing the spread and central tendency of the data between categories, providing insights into the variability and skewness within each group.

```python
sns.catplot(x = 'Category', y = 'Profit', kind = 'box', data = df,  palette = 'hsv', legend = False)
plt.show()
```

```python
sns.boxplot(x = 'Market', y = 'Profit', data = df, palette = 'Set1')
```

**5. Violin Plot**

* A violin plot combines aspects of both box plots and kernel density plots. It displays the distribution of a continuous variable across different categories, showing the probability density of the data at different values.
* The peaks in the plots describes the presence of outliers in the columns.

```python
plt.figure(figsize = (10,8))
sns.catplot(x = 'Sub-Category', y = 'Sales', kind = 'violin', data = df, palette = 'hsv')
plt.xticks(rotation = 60)
```

```python
sns.catplot( x = 'Region', y = 'Shipping Cost', kind  = 'violin', data = df, palette = 'rainbow')
plt.xticks(rotation = 90)
```

```python
sns.catplot(x='Category', y='Profit', kind='violin', data=df, palette='Set1', hue='Category')
```

**9. lmplot**

* The lmplot in Seaborn is a high-level function used to create scatter plots along with a linear regression line, often accompanied by confidence intervals. It is useful for visualizing the relationship between two continuous variables, with an optional categorical variable.

* The "fit_reg" in the code allows you to add linear regression line on the data column and if we keep it "True" the line will be visible and if it is "False" the line will not be visible.

```python
sns.lmplot(x = 'Discount', y = 'Sales', hue = "Category", data = df2, fit_reg = True)
```

```python
sns.lmplot( x= 'Sales', y ='Shipping Cost', hue = 'Sub-Category', data = df2, fit_reg = False)   
```

```python
sns.lmplot( x = 'Discount', y = 'Profit', data = df2, hue = 'Order Priority', fit_reg = True)
```

**10. Scatterplot**

* A scatter plot is a type of data visualization that displays individual data points on a two-dimensional plane, where each point represents a pair of values from two continuous variables. It is useful for showing the relationship or correlation between these variables, identifying trends, patterns, and outliers.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 10))

sns.scatterplot(
    x='Sales',
    y='Discount',
    size=df['Profit'].values,  # Map size to 'Profit'
    sizes=(10, 100),  # Adjust min and max size for better visualization
    hue='Category',  # Map color to 'Category'
    data=df
)

plt.title("Scatter Plot of Sales vs Discount by Category")
plt.xlabel("Sales")
plt.ylabel("Discount")
plt.legend(title="Category")
plt.show()
```

```python
plt.figure(figsize = (10,10))
sns.scatterplot(x = 'Sales', y = 'Profit', hue = 'Market', data = df)
```

```python
plt.figure(figsize = (10,10))
sns.scatterplot(x = 'Discount',
                y = 'Profit',
                hue = 'Market',
                data = df)
```

```python
df.plot(kind='scatter',x='Sales',y='Discount',s=df['Profit'].values, color ='green')
df.plot
```

**11. Heatmap**

* A heatmap is a data visualization technique used to display a matrix of values where individual values are represented by colors. It is especially useful for visualizing correlation matrices, relationships between variables, or frequency distributions in a grid-like format.
* Heatmaps provide an intuitive way to spot patterns, trends, and anomalies in large datasets.

```python
import numpy as np
df3 = df1[['Sales', 'Profit', 'Discount']]
```

```python
df4 = df[['Sales','Profit', 'Discount','Shipping Cost', 'Quantity']]
df4_corr = df4.corr()
sns.heatmap(data = df4_corr, cmap = "rainbow", annot = True)
```

```python
plt.figure(figsize = (50, 30))
sns.heatmap(data = df3, cmap = 'coolwarm')
```
