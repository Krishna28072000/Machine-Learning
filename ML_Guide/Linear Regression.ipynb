### **Linear Regression in ML**


- Linear regression is a statistical method used to model the relationship between a dependent variable (target) and one or more independent variables (predictors) by fitting a linear equation to observed data. 
- The goal is to find the best-fitting line that minimizes the sum of squared differences between the observed values and the predicted values. 
-  It can be used for predicting outcomes and understanding relationships between variables. If thereâ€™s only one predictor, it's called simple linear regression; with multiple predictors, it's multiple linear regression.
- In this file we will look how to perform simple linear regression on a dataset.

**1. Import Necessary Libraries**

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

**2. Load and Read the dataset**

```python
data = pd.read_csv("E:\\Machine Learning\\Housing.csv")
```

```python
data.head()
```

**3. Basic Analysis of data**

```python
data.shape
```

```python
data.isnull().sum()
```

**4. Extracting essential columns**

```python
house_size = data['area'].values
house_prices = data['price'].values
```

**5. Visualization**

```python
plt.scatter(house_size, house_prices, marker = 'o', color = 'blue')
plt.title('HousePrice v/s HouseArea plot')
plt.xlabel('HouseSize (sqft)')
plt.ylabel('HousePrice ($)')
plt.show()
```

**6. Performing Linear Regression**

```python
x_train, x_test, y_train, y_test = train_test_split(house_size, house_prices, test_size=0.2, random_state=42)
```

```python
x_train = x_train.reshape(-1, 1)
x_test = x_test.reshape(-1, 1)

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
model.score(x_train, y_train)
```

```python
# Predict prices for the test set
predictions = model.predict(x_test)

# Visualize the predictions
plt.scatter(x_test, y_test, marker='o', color='blue', label='Actual Prices')
plt.plot(x_test, predictions, color='red', linewidth=2, label='Predicted Prices')
plt.title('Dumbo Property Price Prediction with Linear Regression')
plt.xlabel('House Size (sq.ft)')
plt.ylabel('House Price (millions $)')
plt.legend()
plt.show()
```

```python
r2_score(y_test, predictions)
```
