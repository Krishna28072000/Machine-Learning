**1. "Query" Function**

* The pandas.DataFrame.query() method allows you to query a DataFrame using a string expression. It provides a concise and intuitive way to filter rows based on column values.

```python
df.query('Profit > 1000 & Discount < 0.2')
```

```python
df.query ('Sales > 2000 & Discount < 0.2')
```

```python
df.query('Profit > 1000 & Discount <0.2 & Segment =="Consumer"')
```

```python
df.query('Profit > 1000 & Discount < 0.2 & Segment in["Consumer"]')
```

```python
df.query('Category == "Technology" & City in ["Prato","Jizan"]')
```

```python
df.query('Sales > 2000 & Discount == 0.2 & Category == "Office Supplies"')
```

**2. "Apply" Function**

* The pandas.DataFrame.apply() function allows you to apply a function along an axis of the DataFrame (either rows or columns). It is versatile and can handle row-wise or column-wise operations, as well as more complex transformations.

```python
df['Actual_Cost'] = df.apply(lambda row: row['Sales'] + row['Shipping Cost'], axis=1)
```

```python
df['Actual_Cost']
```

```python
df['Actual_Cost1'] = df.apply(lambda row: row['Profit'] - row['Shipping Cost'], axis=1)
```

```python
df['Actual_Cost1']
```

```python
df['Product Cost'] = df.apply(lambda row : row['Sales'] / row['Quantity'],axis=1)
```

```python
df['Product Cost']
```

**3. "Groupby" Function**

* The pandas.DataFrame.groupby() function is used to group data in a DataFrame based on one or more keys, and then apply an aggregate function or transformation on each group. This is particularly useful for summarizing or analyzing data grouped by specific columns or categories.

* In this we take one categorical and numerical column from the dataset and can understand the relation between them.

```python
df.groupby('State')['Sales'].mean()
```

```python
df.groupby('City')['Profit'].sum()
```

```python
df.groupby('Segment')['Sales'].sum()
```

```python
df.groupby('Product Name')['Sales'].mean()
```

```python
df.groupby('State')['Quantity'].sum()
```

```python
df.groupby('Order Priority')['Shipping Cost'].sum()
```

```python
df.groupby('Category')['Quantity'].max()
```

```python
df.groupby('Ship Mode')['Quantity'].sum()
```

**4. "Making Bins" Function**

* The pandas.cut() function is used to segment and sort data values into discrete intervals, which is particularly helpful for grouping continuous data into bins or categories.

```python
df['Profit Group'] = pd.cut(df['Profit'], bins=[0, 500, 1000, 10000], labels=['Low', 'Medium', 'High'])
```

```python
df['Profit Group']
```

```python
df['Discount Group'] = pd.cut(df['Discount'], bins = [0, 0.25, 0.5, 1], labels = ['Low', 'Medium','High'])
```

```python
df['Discount Group'].unique()
```

### Excercise

```python
#1. Findout which city is having minimum profit

city_profit = df.groupby("City")["Profit"].sum()
min_profit_state = city_profit.idxmin()
min_profit = city_profit.min()
print(f"The state with minimum profit is :{min_profit_state}")
print(f"The state with minimum profit of :{min_profit}") 
```

```python
# 2. Which state has maximum profit
state_profit = df.groupby("State")["Profit"].sum()
max_profit_state = state_profit.idxmax()
max_profit = state_profit.max()
print(f"The state with maximum profit is :{max_profit_state}")
print(f"The state with maxiumum profit of:{max_profit}")
```

```python
# 3. which segment is having maximum sales 
segment_sales = df.groupby("Segment")["Sales"].sum()
max_sales = segment_sales.max()
max_sales_segment = segment_sales.idxmax()
print(f"The segment with maximum sales is :{max_sales_segment}")
print(f"The segment with maximum sales of : {max_sales}")
```

```python
# 4. who is the most repetitive customer
customer_counts = df["Customer Name"].value_counts()
customer = customer_counts[customer_counts>1]
print("The repetitive customer is :}")
print(customer)
```

```python
# 5. which country has maximum shipping cost
max_shipping_cost_country = df.groupby("Country")["Shipping Cost"].sum()
max_country = max_shipping_cost_country.max()
max_shippingcost = max_shipping_cost_country.idxmax()
print(f"max_shipping_cost_country of:{max_country}")
print(f"max_shipping_cost of country is : {max_shippingcost}")
```

```python
# 6. which segment has maximum shipping cost 
max_segment = df.groupby("Segment")["Shipping Cost"].sum()
segment = max_segment.idxmax()
Shipping_cost = max_segment.max()
print(f"Segment with maximum shipping cost is:{segment}")
print(f"Segment with maximum shipping cost of:{Shipping_cost}")
```

```python
# 10. Which region is more profitable ?
region_profit = df.groupby("Region")["Profit"].sum()
region = region_profit.idxmax()
profit = region_profit.max()
print(f"The region which is more profitable is :{region}")
print(f"The region which is more profitable is of :(${profit})")
```

```python
# 11. which region has maximum shipment cost ?
region_cost = df.groupby('Region')['Shipping Cost'].sum()

max_cost_region = region_cost.idxmax()
max_cost_value = region_cost.max()

print(f"Region with maximum shipment cost: {max_cost_region} (${max_cost_value})")
```

```python
# 12. which market has maximum sales?
market_sales = df.groupby("Market")["Sales"].sum()
market = market_sales.idxmax() 
sales = market_sales.max()     

print(f"The market with maximum sales is : {market}")
print(f"The market has sales of amount: {sales}")
```

```python
# 13. Which market is having maximum discount ?
market_discount = df.groupby("Market")["Discount"].sum()
market = market_discount.idxmax()
discount = market_discount.max()
print(f"The market which has maximum discount is:{market} and of dicount amount(${discount})")
```

```python
# 14. Which region has minimum sales ?
minimum_sales = df.groupby("Region")["Sales"].sum()
region = minimum_sales.idxmin()
sales = minimum_sales.min()
print(region)
print(sales)
```

```python
# 15. which category ans subcategory is having maximum discount
max_discount_row = df.loc[df['Discount'].idxmax()]

max_category = max_discount_row['Category']
max_subcategory = max_discount_row['Sub-Category']
max_discount = max_discount_row['Discount']

print(f"The category and subcategory with the maximum discount is: {max_category} - {max_subcategory} with a discount of {max_discount}%.")
```

```python
# 16. which category has mean profit 
category_profit =df.groupby("Category")["Profit"].sum()

profit = category_profit.mean()

print(f"The category with mean profit {profit}")
closest_category = (category_profit -profit).abs().idxmin()
print(f"The category closest to the mean profit is: {closest_category}")

```
