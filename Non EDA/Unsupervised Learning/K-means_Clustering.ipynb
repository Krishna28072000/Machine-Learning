### **K-Means Clustering Algorithm**

* K- Means Clustering is an unsupervised learning algorithm that is used to solve the clustering problems in machine learning or data science.
* It groups the unlabeled dataset into different clusters , Here K defines the number of pre defined clusters that is need to be created in the process.
* It allows us to cluster the data into different groups of an unlabeled dataset on its own without any need for any training.

* It is a centroid based algorithm , where each cluster is associated with a centroid, the main aim of this algorithm is to minimize the sum of distances between the datapoint and their corresponding clusters.

**Step 1 : Import Necessary Libraries**

* Import the libraries for our model, which is a part of data preprocessing. 

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.cluster import KMeans
```

**Step 2 : Import Dataset**

* Import the dataset that we need to use. Here, the dataset we are using is Mall_Customers_data.csv dataset.

```python
df = pd.read_csv("E:\\Machine Learning\\Datasets\\Mall_Customers.csv")
df.head()
```

```python
df.shape
```

**Step 3 : Extracting Independent Variables**

* As we can see the we extracted the third and fourth feature it is because we need a 2-D plot to visualize the model and some features are not requires , suchas customer_ID. 

```python
x = df.iloc[:, [3,4]].values
```

**Step 4 : Determining Number of clusters using Elbow Method**

* The provided code is used for determining the optimal number of clusters in a dataset for K- Means clustering by employing Elbow method.
* Here, "sse" is an empty dictionary to store the inertia (sum of squared distances of samples to their closest cluster center) for different vaslues of "k" (no. of clusters).

* "kmeans.inertia_" stores the inertia (sse) , which is the sum of squared distances of each point to its assigned cluster center.

```python
sse = {}
for k in range(1,10):
    kmeans = KMeans(n_clusters = k, max_iter = 1000).fit(x)
    sse[k] = kmeans.inertia_
```

```python
plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++',random_state = 42)
y_predict = kmeans.fit_predict(x)  
```

```python
#visulaizing the clusters  
plt.scatter(x[y_predict == 0, 0], x[y_predict == 0, 1], s = 100, c = 'blue', label = 'Cluster 1') #for first cluster  
plt.scatter(x[y_predict == 1, 0], x[y_predict == 1, 1], s = 100, c = 'green', label = 'Cluster 2') #for second cluster  
plt.scatter(x[y_predict== 2, 0], x[y_predict == 2, 1], s = 100, c = 'red', label = 'Cluster 3') #for third cluster  
plt.scatter(x[y_predict == 3, 0], x[y_predict == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4') #for fourth cluster  
plt.scatter(x[y_predict == 4, 0], x[y_predict == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5') #for fifth cluster  
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroid')   
plt.title('Clusters of customers')  
plt.xlabel('Annual Income (k$)')  
plt.ylabel('Spending Score (1-100)')  
plt.legend()  
plt.show()  
```
